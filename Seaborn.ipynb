```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# statistica data
# replot Used for creating relational plots
a=sns.load_dataset("flights")
a.columns
```

```python
sns.relplot(x="passengers",y="month",data=a)
```

```python
a=sns.load_dataset("flights")
sns.relplot(x="passengers", y="month",hue="year",data=a)
```

```python
b=sns.load_dataset("tips")
sns.relplot(x="time", y="tip",data=b, kind="line")
```

```python
# categorical data
# catplot used for creating categorical plots
sns.catplot(x="day", y="total_bill",data=b )
```

```python
sns.catplot(x="day", y="total_bill", kind="violin", data=b)
```

```python
sns.catplot(x="day", y="total_bill", kind="boxen",data=b)
```

```python
# Displot used for creating distribution plots
# univarient
c=np.random.normal(loc=5, size=100,scale=2)
sns.displot(c)
```

```python
a=sns.load_dataset("iris")
# FaceGrid
class helps in visualizing distribution of one variable as well as the relationship between multiple variables seperately within subsets of your datasets using multiple 
b=sns.FacetGrid(a,col="species")
b.map(plt.hist, "sepal_length")
```

```python
a=sns.load_dataset("flights")
# This class map each variable in a dataset onto a column and row in a grid of multiple axes.
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style="darkgrid")
a=sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style="dark")
a=sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style="white",color_codes=True)
a=sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill",data=a)
```

```python
# remove axis line
sns.set(style="white",color_codes=True)
a=sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=a)
sns.despine(offset=10,trim=True)
```
